## Conversation Cookbook

This notebook shows common recipes for working with phidata conversations

### Basic conversation

```python
from phi.conversation import Conversation

conversation = Conversation()
conversation.print_response("Share a quick healthy breakfast recipe.")
```

### Adding tools to a conversation

To add tools to a conversation:

1. Import the tools
2. Add it to the conversation

```python
from phi.tools.shell import ShellTools
from phi.conversation import Conversation

shell_conversation = Conversation(tools=[ShellTools()])

shell_conversation.print_response("Show me the files in this directory and their size", stream=False)
```
